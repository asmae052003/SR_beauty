```python
import pandas as pd

reviews_path = "/content/recommendation-engine/data/raw/All_Beauty_25.json.gz"
meta_path    = "/content/recommendation-engine/data/raw/meta_All_Beauty_25.json.gz"

reviews = pd.read_json(reviews_path, compression="gzip")
meta    = pd.read_json(meta_path, compression="gzip")

print("Reviews shape :", reviews.shape)
print("Meta shape    :", meta.shape)
```

```python
meta.columns
```

```python
meta["image"].head()
```

```python
meta_with_image = meta[
    meta["image"].apply(lambda x: isinstance(x, list) and len(x) > 0)
]

print("Items avec image :", meta_with_image.shape)
```

```python
from IPython.display import Image, display

row = meta_with_image.iloc[0]

print("ASIN :", row["asin"])
print("Titre :", row["title"])
print("Brand :", row["brand"])

display(Image(url=row["image"][0]))
```
